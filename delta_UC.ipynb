```python
import configparser
config = configparser.ConfigParser()
config.read("/lakehouse/default/Files/KV/variable.ini")
uc               = config.get("myvars", "uc")
uc_endpoint      = config.get("myvars", "uc_endpoint")
```

```python
from daft.unity_catalog import UnityCatalog
import daft

unity = UnityCatalog(
    endpoint= uc_endpoint,
    token=uc,
)
```

```python
unity.list_tables("databricks_ne.aemo")
```

```python
unity_table = unity.load_table("databricks_ne.aemo.scada")
df = daft.read_deltalake(unity_table)
df.show()
```
